```python
import tkinter as tk
from tkinter import messagebox
from typing import List, Tuple
from graphviz import Digraph
import subprocess

class PDA:
    def _init_(self, states, stack_alphabet, transition_function, start_state, start_stack_symbol, final_states):
        self.states = states
        self.stack_alphabet = stack_alphabet
        self.transition_function = transition_function
        self.start_state = start_state
        self.start_stack_symbol = start_stack_symbol
        self.final_states = final_states
        self.stack: List[str] = []  # Initialize with an empty stack


    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            raise IndexError("pop from an empty stack")

    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        else:
            return None

    def is_empty(self):
        return len(self.stack) == 0
    def visualize_pda(pda):
        dot = Digraph(comment='PDA')

         # Add states
        for state in pda.states:
             dot.node(state, shape='circle', color='blue' if state in pda.final_states else 'black')

         # Add transitions
        for transition, target in pda.transition_function.items():
            start_state, char, stack_symbol = transition
            end_state, *stack_symbols = target

            label = f'{char}, {stack_symbol} -> {",".join(stack_symbols)}'

            dot.edge(start_state, end_state, label=label)

         # Save the image to a file (e.g., pda_graph.png)
        dot.render('pda_graph', format='png', cleanup=False)


def infix_to_postfix(infix_expression: str, pda: PDA) -> Tuple[str, List[str]]:
    output = []
    steps = []

    pda.push(pda.start_stack_symbol)  # Initialize the stack with the start stack symbol
    current_state = pda.start_state

    for char in infix_expression:
        if char.isalnum():
            output.append(char)
            steps.append(f"Push({char})")
        elif char == '(':
            pda.push(char)
            steps.append(f"Push({char})")
        elif char == ')':
            while not pda.is_empty() and pda.peek() != '(':
                output.append(pda.pop())
                steps.append("Pop()")
            pda.pop()
            steps.append("Pop()")
        else:
            while not pda.is_empty() and pda.peek() in pda.stack_alphabet and \
                    (current_state, char, pda.peek()) in pda.transition_function:
                output.append(pda.pop())
                steps.append("Pop()")
            pda.push(char)
            steps.append(f"Push({char})")

    while not pda.is_empty():
        popped_item = pda.pop()
        if popped_item != pda.start_stack_symbol:
            output.append(popped_item)
            steps.append("Pop()")

    return ''.join(output), steps


# Define the PDA tuples
states = {'q0', 'q1', 'q2'}
stack_alphabet = {'0', '1', '(', ')'}
transition_function = {('q0', '0', 'Z0'): ('q1', 'Z0'), ('q1', '0', 'Z0'): ('q1', '0Z0'),
                       ('q1', '1', 'Z0'): ('q2', ''), ('q2', '1', '0'): ('q2', '')}
start_state = 'q0'
start_stack_symbol = 'Z0'
final_states = {'q2'}

pda = PDA(states, stack_alphabet, transition_function, start_state, start_stack_symbol, final_states)

class App:
    def _init_(self, root):
        self.root = root
        self.root.title("Infix to Postfix Converter")

        self.label = tk.Label(root, text="Enter Infix Expression:")
        self.label.pack()

        self.entry = tk.Entry(root)
        self.entry.pack()

        self.convert_button = tk.Button(root, text="Convert", command=self.convert_infix_to_postfix)
        self.convert_button.pack()

        self.result_label = tk.Label(root, text="")
        self.result_label.pack()
        
    def convert_infix_to_postfix(self):
        infix_expression = self.entry.get()

        postfix_expression, steps = infix_to_postfix(infix_expression, pda)

        result_text = f"Infix Expression: {infix_expression}\nPostfix Expression: {postfix_expression}\n\nSteps:\n"
    
        for step in steps:
            result_text += step + "\n"

        self.result_label.config(text=result_text)
        messagebox.showinfo("Result", result_text)
        # Open the image file
        image_file = 'pda_graph.png'
        subprocess.Popen(['start', image_file], shell=True)  # Windows

if _name_ == "_main_":
    main_root = tk.Tk()
    main_root.geometry("400x300")  # Set the initial size of the window
    app = App(main_root)
    
    # Visualize the PDA before starting the main loop
    pda_instance = PDA(states, stack_alphabet, transition_function, start_state, start_stack_symbol, final_states)
    pda_instance.visualize_pda()

    main_root.mainloop()
```
