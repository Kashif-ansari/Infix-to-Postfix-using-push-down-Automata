```python
def infix_to_postfix(expr):
    def get_precedence(op):
        precedence = {'+':1, '-':1, '*':2, '/':2, '^':3}
        return precedence[op] if op in precedence else 0

    def is_operator(c):
        return c in ['+', '-', '*', '/', '^']

    output = []
    stack = []

    for char in expr:
        if char == ' ':  # skip whitespace
            continue
        elif char.isalnum():  # Operand
            output.append(char)
        elif char == '(':  # Left parenthesis
            stack.append(char)
        elif char == ')':  # Right parenthesis
            while stack and stack[-1] != '(':
                output.append(stack.pop())
            stack.pop()
        elif is_operator(char):  # Operator
            while stack and get_precedence(char) <= get_precedence(stack[-1]):
                output.append(stack.pop())
            stack.append(char)

    while stack:
        output.append(stack.pop())

    return ''.join(output)

expr = "a+b*c-d/e*f+g"
print(infix_to_postfix(expr))
exp = input("Enter an expression :")
print(infix_to_postfix(exp))
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("pop from an empty stack")

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            raise IndexError("peek from an empty stack")

    def size(self):
        return len(self.items)
```

```python
import tkinter as tk
from tkinter import messagebox

class PDA:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            raise IndexError("pop from an empty stack")

    def is_empty(self):
        return len(self.stack) == 0


def get_precedence(op):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    return precedence[op] if op in precedence else 0


def infix_to_postfix(infix_expression):
    output = []
    stack = PDA()

    for char in infix_expression:
        if char.isalnum():
            output.append(char)
        elif char == '(':
            stack.push(char)
        elif char == ')':
            while not stack.is_empty() and stack.peek() != '(':
                output.append(stack.pop())
            stack.pop()  # Pop the '(' from the stack
        else:
            while not stack.is_empty() and get_precedence(char) <= get_precedence(stack.peek()):
                output.append(stack.pop())
            stack.push(char)

    while not stack.is_empty():
        output.append(stack.pop())

    return ''.join(output)


def visualize_pda(transitions):
    for transition in transitions:
        print(transition)


class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Infix to Postfix Converter")

        self.label = tk.Label(root, text="Enter Infix Expression:")
        self.label.pack()

        self.entry = tk.Entry(root)
        self.entry.pack()

        self.convert_button = tk.Button(root, text="Convert", command=self.convert_infix_to_postfix)
        self.convert_button.pack()

    def convert_infix_to_postfix(self):
        infix_expression = self.entry.get()
        postfix_expression = infix_to_postfix(infix_expression)
        messagebox.showinfo("Result", f"Infix Expression: {infix_expression}\nPostfix Expression: {postfix_expression}")

        # Visualize PDA transitions (you can replace this with a graphical representation)
        transitions = []
        stack = PDA()
        for char in infix_expression:
            if char.isalnum():
                transitions.append(f"Push({char})")
            elif char == '(':
                stack.push(char)
            elif char == ')':
                while not stack.is_empty() and stack.peek() != '(':
                    transitions.append("Pop()")
                stack.pop()  # Pop the '(' from the stack
            else:
                while not stack.is_empty() and get_precedence(char) <= get_precedence(stack.peek()):
                    transitions.append(f"Pop({stack.pop()})")
                stack.push(char)

        while not stack.is_empty():
            transitions.append(f"Pop({stack.pop()})")

        visualize_pda(transitions)


if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```
